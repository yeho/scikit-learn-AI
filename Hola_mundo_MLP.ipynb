**Hola Mundo scikit-learn**
usando Multi-layer Perceptron (MLP)

```python
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
import matplotlib.pyplot as plt
X, y = make_moons(n_samples=10000, noise=0.2)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
mlp = MLPClassifier(hidden_layer_sizes=(60, 40,20), max_iter=1000)
mlp.fit(X_train, y_train)
accuracy = mlp.score(X_test, y_test)
print(f"Precisión en prueba: {accuracy:.2%}")
```

grafica que muestra el MLP

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(8, 6))

# Plotting the training data points
ax.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap=plt.cm.viridis, s=20, label='Datos de Entrenamiento', alpha=0.6)

# Plotting the test data points
ax.scatter(X_test[:, 0], X_test[:, 1], c=y_test, cmap=plt.cm.viridis, s=40, marker='*', label='Datos de Prueba', alpha=0.8)

# Adding a title that includes the accuracy
ax.set_title(f'Relación entre Datos de Entrenamiento, Prueba y Precisión ({accuracy:.2%})')
ax.set_xlabel('Característica 1')
ax.set_ylabel('Característica 2')
ax.legend()
ax.grid(True)
plt.show()
```
